# Directional Change #

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from IPython import display
from ipywidgets import interact, widgets

%matplotlib inline

data = pd.read_csv('coinbaseUSD_1-min_data_2014-12-01_to_2018-06-27.csv')

data['timestamp'] = pd.to_datetime(data['Timestamp'],unit='s')
data.drop('Timestamp',axis=1,inplace = True)
data['Date'] = data.timestamp.dt.date
data = data.set_index('timestamp');
```

```python
def plot(year,month,day,freq):
    data.loc[str(year)+' '+ month +' '+ str(day)].resample(freq,closed='right',label='right')['Weighted_Price'].mean().plot(figsize=(16,5));
    data.loc[str(year)+' '+ month +' '+ str(day)]['Weighted_Price'].plot(figsize=(16,5));
    sample = data.loc[str(year)+' '+ month +' '+ str(day)].resample(freq,closed='right',label='right').mean().index
    for xc in sample:
        plt.axvline(x=xc,c='k',linewidth=0.5,linestyle='--')
    plt.axis(xmin=sample[0] , xmax= sample[-1]);
```

```python
interact(plot, year = widgets.IntSlider(min=2014,max=2018,step=1,value=2016),month = widgets.SelectionSlider(options=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec'], value='Jan',), day = widgets.IntSlider(min=1,max=30,step=1,value=3),freq =widgets.Text(value='H'))
```
