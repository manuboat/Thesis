```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
PG = pd.read_csv('PG.csv')
```

```python
PG.head()
```

## Simple Return

```python
PG['simple_return'] = (PG['Adj Close']/PG['Adj Close'].shift(1))-1
```

```python
PG
```

```python
PG['simple_return'].plot(figsize=(8,5));
```

```python
daily_avg_returns = PG['simple_return'].sum()
daily_avg_returns
```

```python
year_avg_returns = PG['simple_return'].mean() * 250
year_avg_returns
```

```python
print(str(round(year_avg_returns,5)*100) +' %')
```

## Logaritmic Return

```python
PG['log_return'] = np.log(PG['Adj Close']/PG['Adj Close'].shift(1))
PG['log_return']
```

```python
PG['log_return'].plot(figsize=(8,5));
```

```python
daily_avg_returns = PG['log_return'].mean()
daily_avg_returns
```

```python
year_avg_returns = PG['log_return'].mean() * 250
year_avg_returns
```

```python
print(str(round(year_avg_returns,4)*100) +' %')
```
